```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim

# Load a different example image
image = color.rgb2gray(data.astronaut())  # Convert to grayscale

# Wavelet transform of the image
wavelet = 'bior1.3'
coeffs2 = pywt.wavedec2(image, wavelet, level=2)
LL, details = coeffs2[0], coeffs2[1:]

# Fixed thresholding
threshold = 0.1
coeffs2_thresh = [LL] + [
    (pywt.threshold(detail[0], threshold), pywt.threshold(detail[1], threshold), pywt.threshold(detail[2], threshold))
    for detail in details
]

# Reconstruct the image from the thresholded coefficients
reconstructed_image_standard = pywt.waverec2(coeffs2_thresh, wavelet)

# Ensure that both images are of the same dtype
image = image.astype(np.float64)
reconstructed_image_standard = reconstructed_image_standard.astype(np.float64)

# Calculate PSNR and SSIM
psnr_standard = psnr(image, reconstructed_image_standard, data_range=image.max() - image.min())
ssim_standard = ssim(image, reconstructed_image_standard, data_range=image.max() - image.min())

print(f"Standard Wavelet Compression - PSNR: {psnr_standard:.2f}, SSIM: {ssim_standard:.2f}")

# Display the original and compressed images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(reconstructed_image_standard, cmap='gray')
axes[1].set_title("Standard Compressed Image")
axes[1].axis('off')

plt.show()

# Function to keep only the largest k coefficients
def keep_largest_coefficients(coeffs, k):
    flat_coeffs = np.concatenate([c.flatten() for c in coeffs])
    flat_coeffs_abs = np.abs(flat_coeffs)
    if len(flat_coeffs) <= k:
        return coeffs
    threshold = np.sort(flat_coeffs_abs)[-k]
    new_coeffs = [np.where(np.abs(c) >= threshold, c, 0) for c in coeffs]
    return new_coeffs

# Reconstructing with different numbers of coefficients
coeff_counts = [300,500]
fig, axes = plt.subplots(1, len(coeff_counts) + 1, figsize=(18, 6))

axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

for i, coeff_count in enumerate(coeff_counts):
    # Keep only the largest coeff_count coefficients
    coeffs2_flat = [LL] + [item for sublist in details for item in sublist]
    coeffs2_kept = keep_largest_coefficients(coeffs2_flat, coeff_count)

    # Reshape the coefficients back into the appropriate structure
    LL_kept = coeffs2_kept[0]
    details_kept = [
        (coeffs2_kept[1], coeffs2_kept[2], coeffs2_kept[3]),
        (coeffs2_kept[4], coeffs2_kept[5], coeffs2_kept[6])
    ]

    # Reconstruct the image from the kept coefficients
    reconstructed_image_kept = pywt.waverec2([LL_kept] + details_kept, wavelet)
    reconstructed_image_kept = reconstructed_image_kept.astype(np.float64)

    # Calculate PSNR and SSIM
    psnr_kept = psnr(image, reconstructed_image_kept, data_range=image.max() - image.min())
    ssim_kept = ssim(image, reconstructed_image_kept, data_range=image.max() - image.min())

    axes[i+1].imshow(reconstructed_image_kept, cmap='gray')
    axes[i+1].set_title(f"Restored with {coeff_count} Coeffs\nPSNR: {psnr_kept:.2f}, SSIM: {ssim_kept:.2f}")
    axes[i+1].axis('off')

plt.show()
```

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim, mean_squared_error as mse

# Load an example image
image = data.coins()  # Use a different grayscale image

# Ensure image is of float type for processing
image = image.astype(np.float64)

# Wavelet transform of the image
wavelet = 'bior1.3'
coeffs2 = pywt.wavedec2(image, wavelet, level=2)
LL, details = coeffs2[0], coeffs2[1:]

# Standard Wavelet Compression
threshold = 5  # Adjusted threshold for more compression
coeffs2_thresh = [LL] + [
    (pywt.threshold(detail[0], threshold), pywt.threshold(detail[1], threshold), pywt.threshold(detail[2], threshold))
    for detail in details
]

# Reconstruct the image from the thresholded coefficients
reconstructed_image_standard = pywt.waverec2(coeffs2_thresh, wavelet)

# Crop the reconstructed image to match the original image dimensions
reconstructed_image_standard = reconstructed_image_standard[:image.shape[0], :image.shape[1]]

# Ensure that both images are of the same dtype
reconstructed_image_standard = reconstructed_image_standard.astype(np.float64)

# Calculate metrics for standard compression
data_range = image.max() - image.min()
psnr_standard = psnr(image, reconstructed_image_standard, data_range=data_range)
ssim_standard = ssim(image, reconstructed_image_standard, data_range=data_range)
mse_standard = mse(image, reconstructed_image_standard)

# Calculate compression ratio for standard compression
coeffs2_nonzero = sum(np.count_nonzero(detail) for detail in coeffs2_thresh)
compression_ratio_standard = (image.size - coeffs2_nonzero) / image.size

print(f"Standard Wavelet Compression - PSNR: {psnr_standard:.2f}, SSIM: {ssim_standard:.2f}, MSE: {mse_standard:.2f}, CR: {compression_ratio_standard:.2f}")

# Enhanced Wavelet Compression
def adaptive_threshold(data, method='universal', value=0):
    if method == 'universal':
        sigma = np.median(np.abs(data)) / 0.6745
        threshold = sigma * np.sqrt(2 * np.log(data.size))
    elif method == 'manual':
        threshold = value
    else:
        raise ValueError("Invalid thresholding method.")
    return pywt.threshold(data, threshold)

coeffs2_thresh = [LL] + [
    (adaptive_threshold(detail[0], 'manual', threshold), adaptive_threshold(detail[1], 'manual', threshold), adaptive_threshold(detail[2], 'manual', threshold))
    for detail in details
]

def quantize(data, step_size):
    return np.round(data / step_size) * step_size

step_size = 0.5  # Adjusted step size for more compression
coeffs2_quantized = [quantize(LL, step_size)] + [
    (quantize(detail[0], step_size), quantize(detail[1], step_size), quantize(detail[2], step_size))
    for detail in details
]

# Reconstruct the image from the quantized coefficients
reconstructed_image_enhanced = pywt.waverec2(coeffs2_quantized, wavelet)

# Crop the reconstructed image to match the original image dimensions
reconstructed_image_enhanced = reconstructed_image_enhanced[:image.shape[0], :image.shape[1]]

# Ensure that both images are of the same dtype
reconstructed_image_enhanced = reconstructed_image_enhanced.astype(np.float64)

# Calculate metrics for enhanced compression
psnr_enhanced = psnr(image, reconstructed_image_enhanced, data_range=data_range)
ssim_enhanced = ssim(image, reconstructed_image_enhanced, data_range=data_range)
mse_enhanced = mse(image, reconstructed_image_enhanced)

# Calculate compression ratio for enhanced compression
coeffs2_nonzero_enhanced = sum(np.count_nonzero(detail) for detail in coeffs2_quantized)
compression_ratio_enhanced = (image.size - coeffs2_nonzero_enhanced) / image.size

print(f"Enhanced Wavelet Compression - PSNR: {psnr_enhanced:.2f}, SSIM: {ssim_enhanced:.2f}, MSE: {mse_enhanced:.2f}, CR: {compression_ratio_enhanced:.2f}")

# Display the compressed images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Standard Compressed Image")
plt.imshow(reconstructed_image_standard, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Enhanced Compressed Image")
plt.imshow(reconstructed_image_enhanced, cmap='gray')
plt.axis('off')

plt.show()

# Bar chart comparison
labels = ['PSNR', 'SSIM', 'MSE', 'Compression Ratio']
standard_values = [psnr_standard, ssim_standard, mse_standard, compression_ratio_standard]
enhanced_values = [psnr_enhanced, ssim_enhanced, mse_enhanced, compression_ratio_enhanced]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, standard_values, width, label='Standard')
bars2 = ax.bar(x + width/2, enhanced_values, width, label='Enhanced')

ax.set_xlabel('Metrics')
ax.set_title('Compression Metrics Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

autolabel(bars1)
autolabel(bars2)

plt.show()
```

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim, mean_squared_error as mse

# Load an example image
image = data.coins()  # Use a different grayscale image
image = image.astype(np.float64)

# List of wavelets to use
wavelets = ['bior1.3', 'haar', 'db1']

# Threshold for standard compression
threshold = 5

# Step size for enhanced compression quantization
step_size = 0.5

# Function for adaptive thresholding
def adaptive_threshold(data, method='universal', value=0):
    if method == 'universal':
        sigma = np.median(np.abs(data)) / 0.6745
        threshold = sigma * np.sqrt(2 * np.log(data.size))
    elif method == 'manual':
        threshold = value
    else:
        raise ValueError("Invalid thresholding method.")
    return pywt.threshold(data, threshold)

# Function for quantization
def quantize(data, step_size):
    return np.round(data / step_size) * step_size

# Function to calculate metrics and display images
def evaluate_compression(image, reconstructed_image, title):
    data_range = image.max() - image.min()
    psnr_value = psnr(image, reconstructed_image, data_range=data_range)
    ssim_value = ssim(image, reconstructed_image, data_range=data_range)
    mse_value = mse(image, reconstructed_image)

    print(f"{title} - PSNR: {psnr_value:.2f}, SSIM: {ssim_value:.2f}, MSE: {mse_value:.2f}")

    # Display the compressed image
    plt.figure(figsize=(6, 6))
    plt.title(title)
    plt.imshow(reconstructed_image, cmap='gray')
    plt.axis('off')
    plt.show()

# Main loop to evaluate different wavelets and compression methods
for wavelet in wavelets:
    print(f"Using wavelet: {wavelet}")

    # Wavelet transform of the image
    coeffs2 = pywt.wavedec2(image, wavelet, level=2)
    LL, details = coeffs2[0], coeffs2[1:]

    # Standard Wavelet Compression
    coeffs2_thresh = [LL] + [
        (pywt.threshold(detail[0], threshold), pywt.threshold(detail[1], threshold), pywt.threshold(detail[2], threshold))
        for detail in details
    ]
    reconstructed_image_standard = pywt.waverec2(coeffs2_thresh, wavelet)
    reconstructed_image_standard = reconstructed_image_standard[:image.shape[0], :image.shape[1]]
    reconstructed_image_standard = reconstructed_image_standard.astype(np.float64)
    evaluate_compression(image, reconstructed_image_standard, f"Standard Compressed Image with {wavelet}")

    # Enhanced Wavelet Compression
    coeffs2_thresh_enhanced = [LL] + [
        (adaptive_threshold(detail[0], 'manual', threshold), adaptive_threshold(detail[1], 'manual', threshold), adaptive_threshold(detail[2], 'manual', threshold))
        for detail in details
    ]
    coeffs2_quantized = [quantize(LL, step_size)] + [
        (quantize(detail[0], step_size), quantize(detail[1], step_size), quantize(detail[2], step_size))
        for detail in details
    ]
    reconstructed_image_enhanced = pywt.waverec2(coeffs2_quantized, wavelet)
    reconstructed_image_enhanced = reconstructed_image_enhanced[:image.shape[0], :image.shape[1]]
    reconstructed_image_enhanced = reconstructed_image_enhanced.astype(np.float64)
    evaluate_compression(image, reconstructed_image_enhanced, f"Enhanced Compressed Image with {wavelet}")

    # Decompression using different numbers of coefficients
    for coeff_count in [20, 50, 100]:
        coeffs_flat = pywt.coeffs_to_array(coeffs2)[0]
        coeffs_flat_sorted = np.sort(np.abs(coeffs_flat).ravel())[::-1]
        threshold = coeffs_flat_sorted[coeff_count]

        coeffs2_thresh_few = [LL] + [
            (pywt.threshold(detail[0], threshold), pywt.threshold(detail[1], threshold), pywt.threshold(detail[2], threshold))
            for detail in details
        ]
        reconstructed_image_few = pywt.waverec2(coeffs2_thresh_few, wavelet)
        reconstructed_image_few = reconstructed_image_few[:image.shape[0], :image.shape[1]]
        reconstructed_image_few = reconstructed_image_few.astype(np.float64)
        evaluate_compression(image, reconstructed_image_few, f"Restored Image with {coeff_count} Coefficients using {wavelet}")

# Bar chart comparison for a selected wavelet
selected_wavelet = 'bior1.3'
coeffs2 = pywt.wavedec2(image, selected_wavelet, level=2)
LL, details = coeffs2[0], coeffs2[1:]

coeffs2_thresh = [LL] + [
    (pywt.threshold(detail[0], threshold), pywt.threshold(detail[1], threshold), pywt.threshold(detail[2], threshold))
    for detail in details
]
reconstructed_image_standard = pywt.waverec2(coeffs2_thresh, selected_wavelet)
reconstructed_image_standard = reconstructed_image_standard[:image.shape[0], :image.shape[1]]
reconstructed_image_standard = reconstructed_image_standard.astype(np.float64)

coeffs2_thresh_enhanced = [LL] + [
    (adaptive_threshold(detail[0], 'manual', threshold), adaptive_threshold(detail[1], 'manual', threshold), adaptive_threshold(detail[2], 'manual', threshold))
    for detail in details
]
coeffs2_quantized = [quantize(LL, step_size)] + [
    (quantize(detail[0], step_size), quantize(detail[1], step_size), quantize(detail[2], step_size))
    for detail in details
]
reconstructed_image_enhanced = pywt.waverec2(coeffs2_quantized, selected_wavelet)
reconstructed_image_enhanced = reconstructed_image_enhanced[:image.shape[0], :image.shape[1]]
reconstructed_image_enhanced = reconstructed_image_enhanced.astype(np.float64)

data_range = image.max() - image.min()
psnr_standard = psnr(image, reconstructed_image_standard, data_range=data_range)
ssim_standard = ssim(image, reconstructed_image_standard, data_range=data_range)
mse_standard = mse(image, reconstructed_image_standard)
psnr_enhanced = psnr(image, reconstructed_image_enhanced, data_range=data_range)
ssim_enhanced = ssim(image, reconstructed_image_enhanced, data_range=data_range)
mse_enhanced = mse(image, reconstructed_image_enhanced)

# Calculate compression ratio for standard compression
coeffs2_nonzero = sum(np.count_nonzero(detail) for detail in coeffs2_thresh)
compression_ratio_standard = (image.size - coeffs2_nonzero) / image.size

# Calculate compression ratio for enhanced compression
coeffs2_nonzero_enhanced = sum(np.count_nonzero(detail) for detail in coeffs2_quantized)
compression_ratio_enhanced = (image.size - coeffs2_nonzero_enhanced) / image.size

# Bar chart comparison
labels = ['PSNR', 'SSIM', 'MSE', 'Compression Ratio']
standard_values = [psnr_standard, ssim_standard, mse_standard, compression_ratio_standard]
enhanced_values = [psnr_enhanced, ssim_enhanced, mse_enhanced, compression_ratio_enhanced]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, standard_values, width, label='Standard')
bars2 = ax.bar(x + width/2, enhanced_values, width, label='Enhanced')

ax.set_xlabel('Metrics')
ax.set_title(f'Compression Metrics Comparison using {selected_wavelet}')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

autolabel(bars1)
autolabel(bars2)

plt.show()
```
